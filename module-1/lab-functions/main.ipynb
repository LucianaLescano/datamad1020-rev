```python
def greatest(arr):
    x = 0
    for element in arr:
        if element > x:
            x = element
    return x
```

## 6. Write a function that returns the factorial of a number.

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. 😉`

```python
def st_dev(arr):
    y = 0
    total = 0
    j = 0
    for element in arr:
        y += element
        total += 1
        mean = y / total
    for i in arr:
        j += (i - mean) ** 2
        h = (j / total) **0.5
    return h
```

```python
listilla = [1,2,3]
st_dev(listilla)
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once.

```python
def pangram(string):
    abc = "abcdefghijklmnopqrstuvwxyz"
    for letters in abc: 
        if letters not in string.lower(): 
            return False
    return True
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

```python
def sort_alpha(string):
    listita = [] 
    words = ""
    for letter in string:
        if letter != ",":
            listita.append(letter)
    listita.sort()
    for letter in listita:
        words += letter + ","
    return words
```

```python
my_string = "una,casa,amarilla"
sort_alpha(my_string)
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.
`Valid special characters: # @ ! $ % & ( ) ^ * [ ] { }`

```python
def check_pass(string):
    abc = "abcdefghijklmnopqrstuvwxyz"
    ABC = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
    special = "#@!$%&()^*[]{}"
    for characters in string:
        if characters in abc:
            pass
        elif characters in ABC:
            pass
        elif characters in special:
            return True
    return False
```

```python
my_pass = "Oveja!*"
check_pass(my_pass)
```

```python
# This will test your function 
test_pass(check_pass)
```
